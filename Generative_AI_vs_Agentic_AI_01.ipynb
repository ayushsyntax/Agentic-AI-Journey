## **AI Agent**

**What it is:**
A **software system** that uses AI (usually generative models) along with logic and tools to **perform tasks automatically**. ([GeeksforGeeks][3])

**Key points**

* **Task performer**: executes a series of steps, uses tools/APIs, interacts with systems.
* **Still bounded**: may be programmed to complete specific task sequences.
* Can â€œreasonâ€ about actions using models but doesnâ€™t necessarily set its own goals.
* Example: automated support assistant that queries a database and updates a ticket. ([Coveo][4])

## **Agentic AI**

**What it is:**
A **higher-level autonomous system** that not only performs tasks but **sets plans, adapts, and pursues goals** across multiple steps **with minimal human input**. ([IBM][5])

**Key points**

* **Autonomous planning and decision-making**
* Can break a big goal into sub-goals, use tools, adapt based on results
* **Multi-step workflows** rather than single prompt and response. ([ThoughtSpot][6])

**Simple summary:** Autonomous AI system that *plans and acts on goals*.

# **Other Common AI Terms**

## **LLM (Large Language Model)**

**What it is:**
A generative AI model trained on huge text datasets to understand and produce language.

* Used as a base for many generative AI tools and agents. ([Quiq][2])

---

## **Autonomy (in AI)**

**Definition:**
The ability of a system to operate **without continuous human direction**.

* Generative AI: low autonomy
* Agentic AI: high autonomy. ([Virtuoso QA][8])

---

## **Tool Use**

**Definition:**
When an AI system calls external software (like APIs, search engines, databases) to complete tasks.

* Agents and agentic systems do this; generative models usually do not. ([Medium][9])

---

## **Reactive vs Proactive AI**

**Reactive AI:** waits for prompts and responds.
**Proactive AI:** sets actions and adapts strategies to reach goals.

* Generative AI is reactive.
* Agentic AI is proactive. ([Virtuoso QA][8])
